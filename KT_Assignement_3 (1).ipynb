**Recursion: Climbing Stairs**

---

*Problem Understanding*

THe problem asks how many unique ways a person can reach the top of a staircase with n steps with each move they climb either 1 or 2 steps. This means the number of ways to reach the top depends on the choices made at each step. For instance, if n = 3, the possible sequences are: 1+1+1, 1+2, and 2+1, giving a total of three distinct ways. The problem is basically asking for the number of possible step combinations that sum exactly to n, where the only allowed increments are 1 or 2. This leads to a recursive relationship very similar to the Fibonacci sequence.

*Inputs, Outputs, and Constraints*

The input is a single integer n, where 1 <= n <= 45, representing the total number of steps needed to reach the top. The output is also an integer, representing the number of distinct ways to climb the staircase following the rule that each move can be either one step or two steps. A key constraint is that the sequence grows exponentially, meaning that recursive methods may become inefficient for larger values of n. However, the given recursive formula correctly captures the mathematical structure of the problem: reaching step n can be done by coming from step n-1 (taking 1 step) or from step n-2 (taking 2 steps). Therefore, the total number of ways to reach step n is the sum of the ways to reach steps n-1 and n-2.

*How to Solve the Problem*

The solution will use a simple recursive definition. The function defines two base cases: if n is 0 or 1, there is exactly one way to climb the stairs, either you are already at the top (n = 0) or you take one single step (n = 1). For any n greater than 1, the function will return the sum of climbStairs(n-1) and climbStairs(n-2). This corresponds directly to the reasoning that to arrive at step n, you must have taken either a single step from step n-1 or a double step from step n-2. Recursively computing both possibilities and summing them produces the total number of distinct paths to reach the top. Once the recursion reaches the base cases for smaller values of n, the results are combined upward to produce the final count. For example, when n = 3, the recursion computes climbStairs(2) and climbStairs(1). climbStairs(2) further breaks into climbStairs(1) + climbStairs(0), both of which return 1. Adding these values yields a total of 3 distinct ways to climb three steps.

*Plan to Solve Problem*

The algorithm will use three pointers named low, mid, and high to divide the array into three regions. The low pointer marks the boundary where all elements before it are known to be 0s. As the algorithm moves through the array, it performs three actions depending on the value at nums[mid]. If the value is 0, it swaps nums[mid] with nums[low], then increments both pointers because the region of sorted 0s has expanded. If the value is 1, no action is needed except advancing mid, because 1s belong in the middle. If the value is 2, it swaps nums[mid] with nums[high] and decreases high. Unlike the 0-case, mid does not move here because the swapped value must also be checked. This three-pointer process ensures all values are placed in the correct region in a single traversal Once the mid pointer passes the high pointer, the array is completely sorted into 0s, 1s, and 2s in the required order. Since the algorithm only swaps elements within the existing array and never allocates new space, it satisfies the in-place constraint. The sorting completes in O(n) time because each element is inspected at most once. For example, when given the input [2,0,2,1,1,0], the algorithm rearranges the values to produce [0,0,1,1,2,2]. Similarly, an input such as [2,0,1] becomes [0,1,2].
